```python
import pandas as pd

# Specify the path to your Excel file
excel_file_path = '/content/complet_intent.xlsx'

# Load the Excel file into a Pandas DataFrame
df = pd.read_excel(excel_file_path)
df = df.fillna("")
df = df.iloc[:15]
df
```

```python
intent_dict = df.to_dict(orient="list")
```

```python
intent_dict[' browser_traffic ']
```

```python
import re
Dict= {}

for intent,texts in intent_dict.items():
  for text in texts:
      text = re.sub(r'[^a-zA-Z\s]', '', text).lower()
      Dict[text]=intent
```

```python
Dict
df = pd.DataFrame(list(Dict.items()), columns=['text', 'intent'])
```

```python
df
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
# Split the data into train, test, and validation sets (e.g., 70% train, 15% test, 15% validation)
train_data, temp_data = train_test_split(df, test_size=0.3, random_state=42)
test_data, val_data = train_test_split(temp_data, test_size=0.5, random_state=42)
# Define file paths for each split
train_file = 'train_data.csv'
test_file = 'test_data.csv'
val_file = 'val_data.csv'

# Save the DataFrames to CSV files
train_data.to_csv(train_file, index=False)
test_data.to_csv(test_file, index=False)
val_data.to_csv(val_file, index=False)
```
