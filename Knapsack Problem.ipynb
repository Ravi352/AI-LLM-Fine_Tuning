```python
def knapsack_0_1(values,weights,capacity):
  n = len(values)
  dp = [[0 for _ in range(capacity+1)] for _ in range(len(values)+1)]
  for i in range(1,n+1):
    for j in range(1,capacity+1):
      if weights[i-1]<=j:
        dp[i][j] = max(dp[i-1][j], dp[i-1][j-weights[i-1]] + values[i-1])
      else:
        dp[i][j] = dp[i-1][j]
  return dp[n][capacity]
```

```python
values = [60, 100, 120]
weights = [10, 20, 30]
capacity = 50
knapsack_0_1(values,weights,capacity)
```
