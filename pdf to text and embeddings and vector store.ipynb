```python
# Replace 'input.csv' with the path to your CSV file
csv_file_path = 'New_data7.csv'

# Replace 'output.txt' with the desired path for the output text file
txt_file_path = 'table.txt'

# Open the CSV file for reading and the text file for writing
with open(csv_file_path, 'r', newline='') as csv_file, open(txt_file_path, 'w') as txt_file:
    # Iterate through each line in the CSV and write it to the text file
    for line in csv_file:
        txt_file.write(line)

print(f"CSV file converted to text and saved as '{txt_file_path}'.")
```
