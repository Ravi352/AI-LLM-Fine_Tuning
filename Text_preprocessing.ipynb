```python
import nltk
```

```python
nltk.download('punkt')
```

```python
paragraph = '''What is computer vision? Computer vision is a field of artificial intelligence (AI) that enables computers and systems to derive meaningful information from digital images, videos and other visual inputs — and take actions or make recommendations based on that information.'''
```

```python
sentances = nltk.sent_tokenize(paragraph)
sentances
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
stemmer = PorterStemmer()
```

```python
nltk.download('stopwords')
```

```python
import re
from nltk import WordNetLemmatizer
```

```python
lemm = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
corpus = []
for i in range(len(sentances)):
  review = re.sub('[^a-zA-Z]',' ',sentances[i])
  review = review.lower()
  review = review.split()
  review = [lemm.lemmatize(word) for word in review if not word in set(stopwords.words("english"))]
  review = " ".join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer()
x = vec.fit_transform(corpus).toarray()
```
