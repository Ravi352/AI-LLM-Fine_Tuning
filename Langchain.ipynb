```python
import os
from langchain.llms import OpenAI
import streamlit as st
from langchain import PromptTemplate
from langchain.chains import LLMChain
from langchain.chains import SequentialChain

openai_key = "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"

## For storing the data into the memory
from langchain.memory import ConversationBufferMemory

os.environ["OPEN_API_KEY"] = openai_key

st.title("Search about any Person")

input_text = st.text_input("Search the topic you want")

## Memory for thr complete conversation

person_memory = ConversationBufferMemory(input_key = 'name',memory_key = 'chat_history')
DOB_memory = ConversationBufferMemory(input_key = 'person',memory_key = 'chat_history')
Des_memory = ConversationBufferMemory(input_key = 'dob',memory_key = 'Description')


## Customize prompt template

first_input_pormpt = PromptTemplate(
    input_variables = ['name'],
    Template = "tell me about {name}"
    )

## open AI LLM
llm = OpenAI(temperature = 0.8)
chain = LLMChain(llm =llm,Prompt = first_input_pormpt,verbose = True,output_key = "Person",memory = person_memory)

second_input_pormpt = PromptTemplate(
    input_variables = ['Person'],
    Template = "When {Person} is born"
    )

## open AI LLM
llm = OpenAI(temperature = 0.8)
chain1 = LLMChain(llm =llm,Prompt = second_input_pormpt,verbose = True,output_key = "DOB",memory=DOB_memory)

## Third Input template
third_input_pormpt = PromptTemplate(
    input_variables = ['DOB'],
    Template = "Mention 5 major changes was done around {DOB} in the world"
    )

## open AI LLM
llm = OpenAI(temperature = 0.8)
chain2 = LLMChain(llm =llm,Prompt = third_input_pormpt,verbose = True,output_key = "Description",memory = Des_memory)


### Instead of writting individual chain we have inbuilted function called SequentialChain for doing such kind of operations

parent_chain = SequentialChain(chains = [chain,chain1,chain2],verbose = True,input_variables=['name'],output_variables = ["person","DOB","Description"])

if input_text:
    st.write(parent_chain({'name':input_text}))
    with st.expander('Persona Name'):
        st.info(person_memory.buffer)

    with st.expander('Major Event'):
        st.info(Des_memory.buffer)

```
