```python
import sqlite3
import pandas as pd

df =pd.read_csv("Cog_data.csv")

## Data cleaning
df.columns = df.columns.str.strip()

connection = sqlite3.connect("Cog_data.db")

## load datafile to database
df.to_sql('Torque Data',connection,if_exists='replace')

connection.close()
```

```python
import sqlite3

# Connect to an existing database file
conn = sqlite3.connect('Cog_data.db')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Execute a SELECT query
cursor.execute('SELECT * FROM "Torque Data" ')

# Fetch the query results
rows = cursor.fetchall()

# Print the results
for row in rows:
    print(row)

# Close the cursor and database connection
cursor.close()
conn.close()
```

```python
db = SQLDatabase.from_uri("sqlite:///./Cog_data.db")
toolkit = SQLDatabaseToolkit(db=db,llm=OpenAI(temperature=0))
agent_executor = create_sql_agent(
    llm=OpenAI(temperature=0),
    toolkit=toolkit,
)
```
