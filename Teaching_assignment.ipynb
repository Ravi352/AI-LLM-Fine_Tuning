```python
def talk(text):
    engine.say(text)
    engine.runAndWait()

def take_command():
    try:
        with sr.Microphone() as source:
            print('listening...')
            voice = listener.listen(source)
            command = listener.recognize_google(voice)
    except:
        pass
    return command

def alphabet_teaching_assistant():
      alphabet = 'abcdefghijklmnopqrstuvwxyz'
      current_index = 0
      command = take_command()
      while current_index < len(alphabet):

          user_input = command

          if user_input == alphabet[current_index]:
              print(f"You said '{alphabet[current_index]}'")
              current_index += 1
          else:
              print(f"Please say the current letter: '{alphabet[current_index]}'")

      print("You've completed the alphabet!")

def months_teaching_assistant():
    months = [
        'january', 'february', 'march', 'april', 'may', 'june',
        'july', 'august', 'september', 'october', 'november', 'december'
    ]
    current_index = 0
    while current_index < len(months):
      command = take_command()
      user_input = command.lower()
      if user_input == months[current_index]:
          print(f"You said '{months[current_index]}'")
          current_index += 1
      else:
          print(f"Please say the current Month: '{months[current_index]}'")

    print("You've completed the months!")

def days_teaching_assistant():
    days = [
        'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'
    ]
    current_index = 0

    while current_index < len(days):
        command = take_command()
        user_input = command.lower()
        if user_input == days[current_index]:
            print(f"You said '{days[current_index]}'")
            current_index += 1
        else:
            print(f"Skipping to the next day: '{days[current_index]}'")

    print("You've completed the days!")



def numbers_teaching_assistant():
    numbers = list(map(str, range(1, 11)))
    current_index = 0
    command = take_command()
    user_input = command.lower()
    while current_index < len(numbers):
        user_input = input(f"Say the number '{numbers[current_index]}': ")

        if user_input == numbers[current_index]:
            print(f"You said '{numbers[current_index]}'")
            current_index += 1
        else:
            print(f"Skipping to the next number: '{numbers[current_index]}'")

    print("You've completed the numbers!")

def run_alexa():
  command = take_command()
  if "alphabets" in command:
    alphabet_teaching_assistant()
  if "numbers" in command:
    numbers_teaching_assistant()
  if "days" in command:
    days_teaching_assistant()
  if "alpha" in command:
    months_teaching_assistant()






while True:
    run_alexa()
```
